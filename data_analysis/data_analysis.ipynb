```python
animes = pd.read_csv("~/Data/anime2.csv").dropna()
scalar = lambda x: (x - np.min(x)) / (np.max(x) - np.min(x))
animes["rating_norm"] = animes[["rating"]].apply(scalar)
animes["members_norm"] = animes[["members"]].apply(scalar)
animes["weight"] = 0.6 * animes["rating_norm"] + 0.4 * animes["members_norm"]


plt.figure(figsize = (18,6))
plt.subplot(131)
sns.distplot(animes["rating"])

plt.subplot(132)
sns.distplot(animes["members"])

plt.subplot(133)
sns.distplot(animes["weight"])
plt.savefig("all_fig.jpg")
plt.show()
```
